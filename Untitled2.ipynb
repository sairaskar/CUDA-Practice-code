```python
%%writefile prog.py

import numpy as np
from numba import cuda

@cuda.jit
def hello_cuda() :
  tx = cuda.threadIdx.x
  bx = cuda.blockIdx.x
  bw = cuda.blockDim.x

  idx = tx + bx * bw

  print("Hello, world! I'm thread", idx)
threads_per_block = 5
blocks_per_grid = 4

hello_cuda[blocks_per_grid, threads_per_block]()


```

```python
%%writefile prog1.py
import numpy as np
from numba import cuda
import hashlib

@cuda.jit
def add_array(a, b, c) :
  i = cuda.grid(1)
  if i < a.size:
    c[i] = a[i] + b[i]

N = 20
a= np.arange(N, dtype=np.float32)
b= np.arange(N, dtype=np.float32)

dev_a = cuda.to_device(a)
dev_b = cuda.to_device(b)
dev_c = cuda.device_array_like(a)

add_array[4,8](dev_a, dev_b, dev_c)

c = dev_c.copy_to_host()
print(c)

```

```python
%%writefile prog3.py
import numpy as np
import time
N = 3000

def initalize_matrix(rows, cols) :
  return np.random.randint(0,10,size=(rows,cols))

def matrix_multiply(A, B) :
  return np.dot(A, B)

A = initalize_matrix(N, N)
B = initalize_matrix(N, N)

start_time = time.time()
C = matrix_multiply(A, B)
end_time = time.time()
cpu_duration_ms = (end_time - start_time) * 1000
print("Time taken by CPU:", cpu_duration_ms, "milliseconds")

```

```python
%%writefile prog4.py
import numpy as np
import math
import time
from numba import cuda, int32
N = 3000
@cuda.jit
def matrix_multiply_kernel(A, B, C) :
  row, col = cuda.grid(2)

  if row < N and col < N:
    s = 0
    for k in range(N):
      s += A[row, k] * B[k, col]
    C[row, col] = s

def main():
  A = np.random.randint(0, 10, size=(N, N), dtype=np.int32)
  B = np.random.randint(0, 10, size=(N, N), dtype=np.int32)
  C = np.zeros((N, N), dtype=np.int32)

  d_A = cuda.to_device(A)
  d_B = cuda.to_device(B)
  d_C = cuda.device_array((N, N), dtype=np.int32)

  threads_per_block = (16, 16)
  blocks_per_grid_x = math.ceil(N / threads_per_block[0])
  blocks_per_grid_y = math.ceil(N / threads_per_block[1])
  blocks_per_grid = (blocks_per_grid_x, blocks_per_grid_y)
  start_time = time.time()
  matrix_multiply_kernel[blocks_per_grid, threads_per_block](d_A, d_B, d_C)
  cuda.synchronize()
  end = time.time()

  C = d_C.copy_to_host()
  print(f"Time taken by GPU: {(end - start_time)*1000:.3f} milliseconds")

if __name__ == "__main__":
  main()
```
